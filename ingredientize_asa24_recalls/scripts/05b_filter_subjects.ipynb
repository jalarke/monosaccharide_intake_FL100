```python
# Filter subjects that have at least 75% of their carbohydrates consumed mapped to glycopedia
```

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 39
```

```python
# Load data
asa24 = pd.read_csv('data/asa24/asa24_ingredients/ingredient_fiber_carb_weights_nndc_120721.csv')
matches = pd.read_csv('data/asa24/asa24_ingredients/ingredient_matches_120721.csv')
foods = pd.read_csv('data/asa24/asa24_ingredients/ingredient_frequency_nndc_120721.csv')
```

```python
matches = matches.rename(columns={'Ingredient description_y':'Ingredient description'})
food_list = matches['Ingredient description'].to_list()
```

```python
asa24_match = asa24[asa24['Ingredient description'].isin(food_list)]
```

```python
asa24_no_match = asa24[~asa24['Ingredient description'].isin(food_list)]
```

```python
asa24_no_match['Ingredient description'].unique() #many of the non
```

```python
asa24_match = pd.merge(asa24_match, matches, on='Ingredient description')
```

```python
asa24_match.rename(columns={'glycan_food':'Simple name'}, inplace=True)
```

```python
asa24_match
```

```python
asa24_match.ingredient_consumed_g.sum()/asa24.ingredient_consumed_g.sum()
```

```python
asa24_match.cal_from_carb.sum()/asa24.cal_from_carb.sum()
```

```python
cal_per_sub = asa24_match.groupby('UserName')['cal_from_carb'].agg(np.sum) / asa24.groupby('UserName')['cal_from_carb'].agg(np.sum)
```

```python
cut80 = cal_per_sub.sort_values(ascending=True)[206:]
```

```python
cut80
```

```python
cut75 = cal_per_sub.sort_values(ascending=True)[163:]
```

```python
cut75 = cut75.index.to_list()
```

```python
asa24_75 = asa24[asa24.UserName.isin(cut75)]
```

```python
asa24_75['Ingredient description'].nunique()
```

```python
asa24_foods = pd.merge(asa24_75, matches, on='Ingredient description')
asa24_foods['Ingredient description'].nunique()
```

```python
asa24_foods = asa24_foods.drop_duplicates(subset='Ingredient description')
asa24_foods = asa24_foods[['Ingredient description', 'frequency', 'note']]
```

```python
asa24_foods.to_csv('data/glycan_library/asa_glycan_foods_subject_filtered_120921.csv')
```
